```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['operator','ac_type']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts().head(7)
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Most {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Most Incident Months
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Convert Date to Month
df['Date'] = pd.to_datetime(df['date'], format='%B %d, %Y', errors='coerce')
df['Month'] = df['Date'].dt.month_name()

# Dataset
top_context = df['Month'].value_counts().head(7)
    
# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Most Incident Months")
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Total Incidents Per Time
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Dataset - Get last word if separate by comma
df = df[df['time'] != "?"] # Remove unknown time
df['time'] = df['time'].str.replace(r'[^0-9:]', '', regex=True) # Remove non number char in time except ":"
df['time'] = df['time'].apply(lambda x: f"{x[:2]}:{x[2:]}" if len(x) == 4 else x) # Tidy up time format
df['time'] = df['time'].apply(lambda x: f"{x[:2]}:00" if ":" in x else x) # Group hour
top_context = df['time'].value_counts().head(7)
    
# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Total Incidents Per Time")
plt.xlabel("Time")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Declaration
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Exploratory Data Analysis (EDA) - Pie Chart Comparison of Total Destinations Per Route
# Dataset
df['total_route'] = df['route'].apply(
    lambda x: x.count('-') + 1 if isinstance(x, str) and '-' in x else None
)
# Remove invalid total route
df = df.dropna(subset=['total_route'])
df['total_route'] = df['total_route'].astype(int)
top_context = df['total_route'].value_counts().head(7)

# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Comparison of Total Destinations Per Route")
plt.xlabel("Total Destinations")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Airplane Incidents Per Decade
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Extract decade from the year
df['decade'] = (df['date'].dt.year // 10) * 10

# Count incidents
incident_counts = df['decade'].value_counts().sort_index()

# Plot Line Chart
plt.figure(figsize=(10, 5))
plt.plot(incident_counts.index, incident_counts.values, marker='o', linestyle='-')
# Line Chart Label
for x, y in zip(incident_counts.index, incident_counts.values):
    plt.text(x, y, str(y), ha='right', va='bottom', fontsize=10, color='black')
plt.xlabel("Decade")
plt.ylabel("Total Incidents")
plt.title("Total Airplane Incidents Per Decade (1930s - 2010s)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Airplane Incidents's Fatalities (Death) Per Decade
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Extract decade from the year
df['decade'] = (df['date'].dt.year // 10) * 10

# Sum fatalities per decade
df['all_fatalities'] = pd.to_numeric(df['all_fatalities'], errors='coerce')
fatalities_per_decade = df.groupby('decade')['all_fatalities'].sum()

# Plot Line Chart
plt.figure(figsize=(10, 5))
plt.plot(fatalities_per_decade.index, fatalities_per_decade.values, marker='o', linestyle='-')
# Line Chart Label
for x, y in zip(fatalities_per_decade.index, fatalities_per_decade.values):
    plt.text(x, y, str(y), ha='right', va='bottom', fontsize=10, color='black')
plt.xlabel("Decade")
plt.ylabel("Total Fatalities (Death)")
plt.title("Total Airplane Incidents's Fatalities (Death) Per Decade (1930s - 2010s)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Military Airplane Incidents's Fatalities (Death) Per Decade
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['operator'] = df['operator'].str.lower()

# Extract decade from the year
df['decade'] = (df['date'].dt.year // 10) * 10

# Sum fatalities per decade for military operator
df.loc[df['operator'].str.contains('military', na=False), 'all_fatalities'] = pd.to_numeric(df['all_fatalities'], errors='coerce')
df_military = df[df['operator'].str.contains('military', na=False)]
fatalities_per_decade = df_military.groupby('decade')['all_fatalities'].sum()

# Plot Line Chart
plt.figure(figsize=(10, 5))
plt.plot(fatalities_per_decade.index, fatalities_per_decade.values, marker='o', linestyle='-')
# Line Chart Label
for x, y in zip(fatalities_per_decade.index, fatalities_per_decade.values):
    plt.text(x, y, str(y), ha='right', va='bottom', fontsize=10, color='black')
plt.xlabel("Decade")
plt.ylabel("Total Fatalities (Death)")
plt.title("Total Military Airplane Incidents's Fatalities (Death) Per Decade (1930s - 2010s)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import calendar

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Airplane Incidents's Fatalities (Death) Per Month
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Extract month from the date
df['month'] = df['date'].dt.month

# Sum fatalities per month
df['all_fatalities'] = pd.to_numeric(df['all_fatalities'], errors='coerce')
fatalities_per_month = df.groupby('month')['all_fatalities'].sum()

# Replace month numbers with month names
month_names = [calendar.month_name[i] for i in fatalities_per_month.index]

# Plot Line Chart
plt.figure(figsize=(10, 5))
plt.plot(month_names, fatalities_per_month.values, marker='o', linestyle='-')
# Line Chart Label
for x, y in zip(month_names, fatalities_per_month.values):
    plt.text(x, y, str(y), ha='right', va='bottom', fontsize=10, color='black')
plt.xlabel("Month")
plt.ylabel("Total Fatalities (Death)")
plt.title("Total Airplane Incidents's Fatalities (Death) Per Month")
plt.xticks(rotation=45) 
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Military vs Civilian Airplane Incidents's Fatalities (Death) Per Decade
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['operator'] = df['operator'].str.lower()
df['all_fatalities'] = pd.to_numeric(df['all_fatalities'], errors='coerce')

# Extract decade from the year
df['decade'] = (df['date'].dt.year // 10) * 10

# Military Data
df_military = df[df['operator'].str.contains('military', na=False)]
fatalities_military = df_military.groupby('decade')['all_fatalities'].sum()

# Civilian Data
df_civilian = df[~df['operator'].str.contains('military', na=False)]
fatalities_civilian = df_civilian.groupby('decade')['all_fatalities'].sum()

# Plot Line Chart
plt.figure(figsize=(10, 5))
# Military Line
plt.plot(fatalities_military.index, fatalities_military.values, marker='o', linestyle='-', color='red', label='Military')
for x, y in zip(fatalities_military.index, fatalities_military.values):
    plt.text(x, y, str(y), ha='right', va='bottom', fontsize=10, color='red')
# Civilian Line
plt.plot(fatalities_civilian.index, fatalities_civilian.values, marker='o', linestyle='-', color='blue', label='Civilian')
for x, y in zip(fatalities_civilian.index, fatalities_civilian.values):
    plt.text(x, y, str(y), ha='right', va='bottom', fontsize=10, color='blue')
# Line Chart Label
plt.xlabel("Decade")
plt.ylabel("Total Fatalities (Death)")
plt.title("Total Military vs Civilian Airplane Incidents's Fatalities (Death) Per Decade")
plt.legend()
plt.grid(True)
plt.show()
```
