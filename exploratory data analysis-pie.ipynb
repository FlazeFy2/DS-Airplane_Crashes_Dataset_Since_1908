```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['operator','ac_type']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts().head(7)
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Most {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Most Incident Months
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Convert Date to Month
df['Date'] = pd.to_datetime(df['date'], format='%B %d, %Y', errors='coerce')
df['Month'] = df['Date'].dt.month_name()

# Dataset
top_context = df['Month'].value_counts().head(7)
    
# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Most Incident Months")
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Total Incidents Per Time
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Dataset - Get last word if separate by comma
df = df[df['time'] != "?"] # Remove unknown time
df['time'] = df['time'].str.replace(r'[^0-9:]', '', regex=True) # Remove non number char in time except ":"
df['time'] = df['time'].apply(lambda x: f"{x[:2]}:{x[2:]}" if len(x) == 4 else x) # Tidy up time format
df['time'] = df['time'].apply(lambda x: f"{x[:2]}:00" if ":" in x else x) # Group hour
top_context = df['time'].value_counts().head(7)
    
# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Total Incidents Per Time")
plt.xlabel("Time")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Exploratory Data Analysis (EDA) - Pie Chart Comparison Between Military and Non-Military Incident
# Dataset
df['Category'] = df['operator'].apply(lambda x: 'Military' if isinstance(x, str) and 'Military' in x else 'Non-Military')
category_counts = df['Category'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
category_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Between Military and Non-Military Incident")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Declaration
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Exploratory Data Analysis (EDA) - Pie Chart Comparison of Total Destinations Per Route
# Dataset
df['total_route'] = df['route'].apply(
    lambda x: x.count('-') + 1 if isinstance(x, str) and '-' in x else None
)
# Remove invalid total route
df = df.dropna(subset=['total_route'])
df['total_route'] = df['total_route'].astype(int)
top_context = df['total_route'].value_counts().head(7)

# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Comparison of Total Destinations Per Route")
plt.xlabel("Total Destinations")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
