```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['operator','ac_type']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts().head(7)
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Most {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Most Incident Months
df = pd.read_csv("Airplane_crashes_dataset_since_1908.csv", encoding="ISO-8859-1")

# Convert Date to Month
df['Date'] = pd.to_datetime(df['date'], format='%B %d, %Y', errors='coerce')
df['Month'] = df['Date'].dt.month_name()

# Dataset
top_context = df['Month'].value_counts().head(7)
    
# Bar Chart
plt.figure(figsize=(10, 6))
ax = top_context.plot(kind='bar')
# Add label
for i, value in enumerate(top_context):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Most Incident Months")
plt.xlabel("Month")
plt.ylabel("Number of Incidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
